```python
#Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
#Import Dataset
data = pd.read_csv("logistic_data.csv")
```

```python
#Load first 05 columns
data.head()
```

```python
#Check the shape of the dataset
data.shape
```

```python
plt.scatter(data.Age,data.Purchased)
```

```python
#Check the blank data in the dataset
data.isnull().sum()
```

```python
#Classify the Axises
x = np.array(data[['Age']])
y = np.array(data[['Purchased']])
```

```python
#Get the dataset data to train & Test
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python
#Check the shape
x_train.shape
```

```python
#Check the shape
x_test.shape
```

```python
#Check the shape
y_train.shape
```

```python
#Check the shape
y_test.shape
```

```python
#Create Model
model = LogisticRegression()
```

```python
#Train the Created Model
model.fit(x_train,y_train)
```

```python
#Check the data
model.predict(x_test)
```

```python
#Check the data
y_test
```

```python
#Check the score of the model predictions
model.score(x_test,y_test)
```

```python
#Check the imput data from the model
age = np.array([[45]])
model.predict(age)
```

```python
#Create a sigmoid data chart to show the dataset accuracy
import numpy as np
import matplotlib.pyplot as plt

# Define the sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate x values
x = np.linspace(-10, 10, 100)

# Calculate y values using the sigmoid function
y = sigmoid(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='Sigmoid Function', color='blue')
# Add scatter points (every 10th point to avoid clutter)
plt.scatter(x[::10], y[::10], color='red', s=50, label='Data Points')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')
plt.grid(True)
plt.legend()
plt.axhline(y=0.5, color='r', linestyle='--', alpha=0.3)
plt.axvline(x=0, color='r', linestyle='--', alpha=0.3)

# Show the plot
plt.show()
```
